## Matplotlib

```python
import matplotlib.pylab as plt
import numpy as np

# array of numbers from -10 - 10 equally spaced between.
x = np.linspace(-10, 10, 20)
f = lambda x: np.cos(x)
```

## MATLAB style interface 

```python
plt.plot(x, f(x), ".--")

```

```python
x = np.linspace(-10, 10, 200)

plt.figure(figsize=(8, 4), dpi=200)
plt.plot(x, f(x))
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.box(False)
```

## OOP approach

```python
fig, ax = plt.figure(figsize=(8, 4)), plt.axes()
ax.plot(x, f(x))  # ax object has a plot method
ax.set(xlabel="x", ylabel="y", title="$\cos{x}$", xlim=(-5, 5), ylim=(-1.2, 1.2))
fig.savefig("../Cos(x).png", dpi=200)

```

## Subplots

```python
plt.figure(figsize=(16, 4))
plt.subplot(1, 2, 1)  # (rows, columns, panel numbers)
plt.plot(x, f(x))
plt.plot(x, np.sin(x))
plt.title("cos(x), sin(x)")


plt.subplot(1, 2, 2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("Trig")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 5))

ax[0].plot(x, f(x))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, f(x))

ax[0].set(title="Cos(x), Sin(x)")

```

## Example plots

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

data_normal = np.random.randn(1000)

ax[0].hist(data_normal, bins=50, alpha=0.7)
ax[0].set(title="Hoistogram of normally distributed data", xlabel="x", ylabel="y")


size = 1000 * np.random.randn(100)

scatter = ax[1].scatter(
    x=np.random.randn(100),
    y=np.random.randn(100),
    s=size,
    alpha=0.6,
    cmap="viridis",
    c=np.random.randn(100),
)

ax[1].set(title= "bubble chart", xlabel= "x", ylabel= "y")
fig.colorbar(scatter)

```
